```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('durations.pkl', 'rb') as f:
    durations = pickle.load(f)
```

# M = 6, S = 3, a = 0.5

```python
plt.plot(range(1, 40, 1)[:len(durations)], durations)
plt.ylabel("Seconds")
plt.xlabel("# Demand Locations")
plt.title("Runtime vs. # Demand Locations (M = 6, S = 3, a = 0.5)")
plt.savefig('growth.png')
```
